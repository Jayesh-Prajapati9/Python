```python
##3
f=open("scores.txt",'r')
d={}
for line in f:
    L=line.split()
    if int(L[2])-int(L[4])>=30:
           d[L[3]]=d.get(L[3],0)+1
    if int(L[4])-int(L[2])>=30:
           d[L[1]]=d.get(L[1],0)+1
##print(d)
k=list(d.items())
k.sort(key=lambda x:x[1],reverse=True)
print(k[0][0],k[0][1])
```

```python
##4
file=open("scores.txt",'r')
d={}
for line in file:
    L=line.split()
    d[L[1]]=d.get(L[1],[])+[int(L[2])]
    d[L[3]]=d.get(L[3],[])+[int(L[4])]
##print(d)
for i,j in d.items():
    if(sum(j)/len(j))>=70:
        print(i,sum(j)/len(j))

```

```python
##5
f=open("scores.txt",'r')
games=[]
d={}
count=0
for line in f:
    L=line.split()
    games.append([L[1],int(L[2]),L[3],int(L[4])])
##print(games)
for team1,score1,team2,score2, in games:
    if team1 not in d:
        d[team1]={'win':0,'loss':0,'my_score':0,'against_score':0}
    if team2 not in d:
        d[team2]={'win':0,'loss':0,'my_score':0,'against_score':0}
    d[team1]['my_score']+=score1
    d[team1]['against_score']+=score2
    d[team2]['my_score']+=score2
    d[team2]['against_score']+=score1
    if score2>score1:
        d[team2]['win']+=1
        d[team1]['loss']+=1
    else:
        d[team1]['win']+=1
        d[team2]['loss']+=1
##print(d)  
for i,j in d.items():
    if(j['win']>j['loss']and(j['my_score']<j['against_score'])):
        print(count,i)
        count+=1
    
    
```
